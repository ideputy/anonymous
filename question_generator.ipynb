## Item 6 question generator

```python
# Moodle Question Template
def filltemp(q_n, text, result):
    return f'''  <question type="numerical">
    <name>
      <text>{q_n}</text>
    </name>
    <questiontext format="html">
      <text><![CDATA[{text}]]></text>
    </questiontext>
    <generalfeedback format="html">
      <text></text>
    </generalfeedback>
    <defaultgrade>1.0000000</defaultgrade>
    <penalty>1.0000000</penalty>
    <hidden>0</hidden>
    <idnumber></idnumber>
    <answer fraction="100" format="moodle_auto_format">
      <text>{result}</text>
      <feedback format="html">
        <text></text>
      </feedback>
      <tolerance>0</tolerance>
    </answer>
    <unitgradingtype>0</unitgradingtype>
    <unitpenalty>0.1000000</unitpenalty>
    <showunits>3</showunits>
    <unitsleft>0</unitsleft>
  </question>'''
```

```python
# Script for character substitution (substitute "start"-symbol with "end"-symbol)
def scriptrepl(start, end):
    return f'''<script type="text/javascript">
    function addLink() {{
        var body_element = document.getElementsByTagName('body')[0];
        var selection;
        selection = window.getSelection();
        var copytext = selection.toString().replace('{start}', '{end}');
        var newdiv = document.createElement('div');
        newdiv.style.position = 'absolute';
        newdiv.style.left = '-99999px';
        body_element.appendChild(newdiv);
        newdiv.innerHTML = copytext;
        selection.selectAllChildren(newdiv);
        window.setTimeout(function() {{
            body_element.removeChild(newdiv);
        }}, 0);
    }}
    document.oncopy = addLink;
</script>'''
```

```python
# Variants for 2-layer NNet
forfc2 = [(2, 7), (4, 5)]

# Variants for 3-layer NNet
forfc3 = [(3, 2), (3, 4)]
```

```python
# Collect of Moodele-formatted variants in the list and enumerate
qslist = []
q_n = 0

# For 2-layer NNet
for feats, hidds in forfc2:
    if feats in [2, 3]:
        start, end = (3, 2) if feats == 3 else (2, 3)
    else:
        start, end = (5, 6) if hidds == 5 else (6, 5)
    textq = f'<p dir="ltr" style="text-align: left;">{{mlang en}}A fully connected two-layer neural network takes {feats} features as input, contains {hidds} neurons in the hidden layer and 1<span style="font-size: 0; color: white">1</span> neuron in the output layer. How many parameters does it have? Enter an integer number (answer only: do not perform a detailed calculation).{{mlang}}</p>'
    result = (feats + 1) * hidds + (hidds + 1) * 1
    qslist.append(filltemp(f'6_{q_n}', textq + '\n' + scriptrepl(start, end), result))
    q_n += 1

# For 3-layer NNet
for feats, hidds in forfc3:
    start, end = (3, 2) if feats == 3 else (2, 3)
    textq = f'<p dir="ltr" style="text-align: left;">{{mlang en}}A fully connected three-layer neural network takes {feats} features as input, contains {hidds} neurons in the hidden layers and 1<span style="font-size: 0; color: white">1</span> neuron in the output layer. How many parameters does it have? Enter an integer number (answer only: do not perform a detailed calculation).{{mlang}}</p>'
    result = (feats + 1) * hidds + (hidds + 1) * hidds + (hidds + 1) * 1
    qslist.append(filltemp(f'6_{q_n}', textq + '\n' + scriptrepl(start, end), result))
    q_n += 1
```

## Item 7 (with invisible spaces)

```python
# Invisible space tag
addspace = '<span style="visibility:hidden;"> </span>'

# Example of template phrases
phrases = f'''What if your pet could only talk{addspace}to you at midnight for an hour
Imagine a funny situation where a{addspace}friend tries to{addspace}copy everything you do to irritate you and write about it
Write about a time in{addspace}your life when you struggled with a{addspace}choice and made the right one'''.splitlines()
```

```python
# Collect Item 7 into list
newqlist = []
q_n = 0

for i in phrases:
    for feats in range(2, 4):
        start, end = (3, 2) if feats == 3 else (2, 3)
        textq = f'<p dir="ltr" style="text-align: left;">{{mlang en}}How many {feats}-grams are present in the following phrase: <strong>{i}</strong><br><br>Enter an integer number (answer only: do not perform a detailed calculation).{{mlang}}</p>'
        result = len(' '.join(re.sub('<.*?>', ' ', i).split()).split()) - (feats - 1)
        newqlist.append(filltemp(f'7_{q_n}', textq + '\n' + scriptrepl(start, end), result))
        q_n += 1
```

## Items 8-9 with leading hidden negation

```python
# Prepare options (some examples)
# First line positive, second line - negative [indicate correctness with "x"]
# Options contain tranlations for multi-language support (delimited with "|")
myoptions = '''[-] A typical GAN trains the generator and the discriminator simultaneously | Обычная GAN обучает генератор и дискриминатор одновременно  
[x] A typical GAN does not train the generator and the discriminator simultaneously | Обычная GAN не обучает генератор и дискриминатор одновременно  

[-] A GAN always uses the same loss function for both discriminator and generator training | GAN всегда использует одну и ту же функцию потерь для обучения дискриминатора и генератора  
[x] A GAN does not always use the same loss function for both discriminator and generator training | GAN не всегда использует одну и ту же функцию потерь для обучения дискриминатора и генератора  

[x] Computers cannot inherently understand ambiguous instructions like humans do | Компьютеры изначально не могут понимать неоднозначные инструкции так, как это делают люди  
[-] Computers inherently understand ambiguous instructions like humans do | Компьютеры изначально могут понимать неоднозначные инструкции так, как это делают люди  

[x] Natural language generation is a subfield of AI that generates texts like how humans write | Генерация естественного языка — это подраздел ИИ, который относится к генерации текстов так, как их пишут люди  
[-] Natural language generation is not a subfield of AI that generates texts like how humans write | Генерация естественного языка не является подразделом ИИ, который относится к генерации текстов так, как их пишут люди  

[-] Natural language generation is a type of Large Language Model that generates images from texts | Генерация естественного языка — это тип больших языковых моделей, позволяющий генерировать изображения из текстов  
[x] Natural language generation is not a type of Large Language Model that generates images from texts | Генерация естественного языка не является типом больших языковых моделей, позволяющий генерировать изображения из текстов  '''
```

```python
# Prepare question database
s7 = pd.DataFrame([i for i in myoptions.splitlines() if i != ''])
s7 = s7.assign(nq=[1,2] * (s7.shape[0]//2)) # for question number ID
s7['TF'] = s7[0].apply(lambda x: False if x[:3] == '[-]' else True)
s7['ENG'] = s7[0].apply(lambda x: x[3:].split('|')[0].strip())
s7['RUS'] = s7[0].apply(lambda x: x[3:].split('|')[1].strip())
s7.iloc[0::2, 1] = range(s7.iloc[0::2, 1].shape[0])
s7.iloc[1::2, 1] = range(s7.iloc[0::2, 1].shape[0])

# Display
s7.head()
```

```python
# Supporting generators for Moodle-format
## Generate multi-language text support
def GENERATE_MUL_TEXT(ENG, RUS):
    return f'''{{mlang en}}
<p dir="ltr" style="text-align: left;">{ENG}</p>
{{mlang}}
{{mlang ru}}
<p dir="ltr" style="text-align: left;">{RUS}</p>
{{mlang}}'''

## One option generator
def GEN_ANSWER(ANSWER, correct=100):
    return f'''<answer fraction="{correct}" format="html">
      <text><![CDATA[{ANSWER}]]></text>
      <feedback format="html">
        <text></text>
      </feedback>
    </answer>'''

## One variant generator
def MULTCHOICE(ID, QUESTION_TEXT, ANSWERS):
    return f'''<question type="multichoiceset">
    <name>
      <text>{ID}</text>
    </name>
    <questiontext format="html">
      <text><![CDATA[<p dir="ltr" style="text-align: left;">{QUESTION_TEXT}</p>]]></text>
    </questiontext>
    <generalfeedback format="html">
      <text></text>
    </generalfeedback>
    <defaultgrade>2.0000000</defaultgrade>
    <penalty>1.0000000</penalty>
    <hidden>0</hidden>
    <idnumber></idnumber>
    <shuffleanswers>true</shuffleanswers>
    <correctfeedback format="html">
      <text></text>
    </correctfeedback>
    <incorrectfeedback format="html">
      <text></text>
    </incorrectfeedback>
    <answernumbering>none</answernumbering>
    <showstandardinstruction>0</showstandardinstruction>
    {ANSWERS}
    <hint format="html">
      <text></text>
      <options>1</options>
    </hint>
  </question>
'''
```

```python
# Generate questions in list
## Random selection from database
## Assign distractor to 1, 2, or 3 options (random)
## Finally two variants of question generated (select all True, and select all False)

# Number of variants to generate (finally * 2)
n_vars = 3
qq8 = []
myid = 0

for hh in range(n_vars):
    dd = s7.sample(frac=1)
    dd = dd.drop_duplicates('nq')
    dd = dd.sample(4)
    pois = dd['TF'].sample(random.randint(1, 3))
    dd = dd.assign(spoil=~pois)
    dd.spoil = dd.spoil.mask(dd.spoil.isna(), dd.TF)
    dd['spoilru'] = dd.spoil.map({True: 'Истина', False: 'Ложь'})
    for is_true in [True, False]:
        QUESTION_TEXT = GENERATE_MUL_TEXT(f'True or False. Identify all {"true" if is_true else "false"} statements:', f'Истина или Ложь. Определите все {"истиные" if is_true else "ложные"} утверждения:')
        ANSZ = []
        for b, qq in dd.iterrows():
            ANSZ.append(GEN_ANSWER(GENERATE_MUL_TEXT('<span style="font-size: 0; color: white">' + str(qq.spoil) + ': </span>' + qq.ENG,
                                    '<span style="font-size: 0; color: white">' + qq.spoilru + ': </span>' + qq.RUS), 100 if qq.TF==is_true else 0))
        if (dd.TF == is_true).sum() == 0: # 100 pts to None answer
            ANSZ.append(GEN_ANSWER(GENERATE_MUL_TEXT('<strong>None of mentioned</strong>', '<strong>Ни одно из упомянутых</strong>')))
        else:
            ANSZ.append(GEN_ANSWER(GENERATE_MUL_TEXT('<strong>None of mentioned</strong>', '<strong>Ни одно из упомянутых</strong>'), 0))
        qq8.append(MULTCHOICE(f'7_{myid}', QUESTION_TEXT, ''.join(ANSZ)))
        myid += 1
```

```python
# Generate curve (may be experimented with parameters)
def gen_curve():
    gap = 10
    j = 10
    a = 'x, y + (i * lineheight),'
    m = 10
    for i in range(1, 16):
        if i % 2 == 0: 
            j+=random.randint(10, 15)
            a+= (f'x + {j}, y + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + {gap//3} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + {gap//3*2} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + {gap} + (i * lineheight),')        
            j+=random.randint(10, 20)
            a+= (f'x + {j}, y + {gap} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + {gap//3*2} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + {gap//3} + (i * lineheight),')
        else:
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y - {gap//3} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y - {gap//3*2} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y - {gap} + (i * lineheight),')
            j+=random.randint(10, 20)
            a+= (f'x + {j}, y - {gap} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y - {gap//3*2} + (i * lineheight),')
            j+=random.randint(10, 30)
            a+= (f'x + {j}, y - {gap//3} + (i * lineheight),')
        if j > 500:
            break

    return a[:-1]
```

```python
# Generate wavy text with line breaks
def canva_hall_q(idnum, ENG, RUS):
    mycurve = gen_curve()
    en_text = f'''{repr(chr(10)).replace("'","")}'''.join(textwrap.wrap(ENG, 50, break_long_words=False))
    ru_text = f'''{repr(chr(10)).replace("'","")}'''.join(textwrap.wrap(RUS, 50, break_long_words=False))
    return f'''<script type="text/javascript" src="https://cdn.rawgit.com/Viglino/Canvas-TextPath/master/ctxtextpath.js"></script>
    <div width="100%">{{mlang en}}
    <canvas id="canvas_{idnum}" width="500" height="{13 + (en_text.count(repr(chr(10)).replace("'","")) + 1) * 20}" style="display:block;width:100%;max-height:200px;max-width:500px;"></canvas>{{mlang}}{{mlang ru}}
    <canvas id="canvas_{idnum}" width="500" height="{13 + (ru_text.count(repr(chr(10)).replace("'","")) + 1) * 20}" style="display:block;width:100%;max-height:200px;max-width:500px;"></canvas>{{mlang}}
</div>
{{mlang en}}<script>
    var canvas = document.getElementById('canvas_{idnum}');
    var context = canvas.getContext('2d');
    context.fillStyle = "rgba(255, 255, 255, 0)";
    context.fillRect(0, 0, 500, 500);
    context.font = "12pt 'Trebuchet MS'";
    context.strokeStyle = 'rgba(0,0,0,.5)';;
    var txt = "{en_text}";
    var x = 10;
    var y = 21;
    var lineheight = 20;
    var lines = txt.split({repr(chr(10))});
    for (var i = 0; i < lines.length; i++)
        context.textPath(lines[i], [{mycurve}]);
</script>{{mlang}}{{mlang ru}}<script>
    var canvas = document.getElementById('canvas_{idnum}');
    var context = canvas.getContext('2d');
    context.fillStyle = "rgba(255, 255, 255, 0)";
    context.fillRect(0, 0, 500, 500);
    context.font = "12pt 'Trebuchet MS'";
    context.strokeStyle = 'rgba(0,0,0,.5)';;
    var txt = "{ru_text}";
    var x = 10;
    var y = 21;
    var lineheight = 20;
    var lines = txt.split({repr(chr(10))});
    for (var i = 0; i < lines.length; i++)
        context.textPath(lines[i], [{mycurve}]);
</script>{{mlang}}'''
```
